```python
import numpy
import networkx
%matplotlib inline

V = list(range(6))
E = [(0, 1), (0, 3), (0, 5), (1, 3), (5, 3), (3, 2), (3, 4),(2,5)]
G = networkx.Graph()
G.add_nodes_from(V)
G.add_edges_from(E)

networkx.draw(G, with_labels=True)


def count_triangles(A, ret_C=False):
    assert (type(A) is numpy.ndarray) and (A.ndim == 2) and (A.shape[0] == A.shape[1])
    ### BEGIN SOLUTION
    C = (A @ A) * A
    n_tri = C.sum() // 6
    if ret_C:
        return n_tri, C
    return n_tri
    ### END SOLUTION

A = networkx.convert_matrix.to_numpy_array(G, dtype=int)
print("The number of triangles are",count_triangles(A))
```
